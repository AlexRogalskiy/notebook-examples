```python
import os
import matplotlib.pyplot as plt
from matplotlib import pyplot
import seaborn as sns
import pandas as pd

print(os.environ['INPUT_DIR'])
print(os.environ['OUTPUT_DIR'])
```

```python
data = pd.read_csv(os.path.join(os.environ['INPUT_DIR'], 'data.csv'))

data['Timestamp'] = pd.to_datetime(data['Timestamp'], unit='ms')
data.set_index('Timestamp', inplace=True)

data.head()
```

```python
plt.figure(figsize=(20,10))
histogram = sns.distplot(data[['internal-temp']])
```

```python
## Save Histogram
fig = histogram.get_figure()

output_path = os.path.join(os.environ['OUTPUT_DIR'], 'histogram.png')
fig.savefig(output_path)
```
